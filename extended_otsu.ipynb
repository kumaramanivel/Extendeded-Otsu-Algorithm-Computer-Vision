```python
import PIL
from PIL import Image
import imageio
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
pic = imageio.imread('tiger1-24bits.bmp')
gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
gray = gray(pic)  
plt.imshow(gray, cmap = plt.get_cmap(name = 'gray'))
im = np.array(Image.open('tiger1-24bits.bmp').convert('L'))
print(type(im))
gr_im= Image.fromarray(im).save('tiger1-24bits_out.bmp')
def otsu_thval(image): #To find TH Value using Otsu method
    hist = histogram(image)
    th = 0
    total_pixel = np.sum(hist)
    levels = np.array(range(0,256))
    temp = -1 
    for i in range(1, len(levels)):
        wb = (np.sum(hist[:i])) / total_pixel # weight of background 
        wf = (np.sum(hist[i:])) / total_pixel # and foreground
        meanb = np.sum(levels[:i] * hist[:i]) / np.sum(hist[:i])
        meanf = np.sum(levels[i:] * hist[i:]) / np.sum(hist[i:])
        bcv = wb * wf * (meanb - meanf) ** 2
        if bcv > temp:
            th = i
            temp = bcv  
    return th
def histogram(image): #OP histogram
    shape = np.shape(image)
    histogram = np.zeros(256)
    if len(shape) == 3:
        image = image.mean(axis=2)
    histogram = np.histogram(image.ravel(), np.array(range(0,257)))[0]
    return histogram # pixel values must be returned to find otsu threshold image
def threshold(image, th=None): #To a binarised version of input.
    shape = np.shape(image)
    binarised = np.zeros([shape[0], shape[1]], dtype=np.uint8)
    if len(shape) == 3:
        image = image.mean(axis=2)
    if th is None:
        th = otsu_thval(image)
        print(f'The threshold value is {th}')
    binarised = image > th
    return binarised
if __name__ == "__main__":
    filename = "tiger1-24bits_out.bmp"
    image = io.imread(filename)
    
    binarised_image = threshold(image)
    fig, axes = plt.subplots(ncols=2, figsize=(10,4))
    ax = axes.ravel()
    ax[0].imshow(image, cmap=plt.cm.gray)
    ax[0].set_title("Original Image")
    
    ax[1].imshow(binarised_image, cmap=plt.cm.gray)
    ax[1].set_title("Binarised Image")
```
